## **Exploratory data analysis** ##

**Importing the libraries for the analysis**

```python
import pandas as pd
import os
import openpyxl
import matplotlib
import seaborn
```

```python
# Loading our dataset
sales = pd.read_csv("sales_data.xls")
```

```python
# Checking the data type for the columns
sales.dtypes
```

```python
# Checking for duplicate data
sales.duplicated().sum
sales
```

```python
# Checking for missing data
sales.isnull().sum
sales
```

```python
# Checking for the size of the dataset
sales.shape
```

```python
# Indexing the dataset
sales1 = pd.DataFrame(sales, columns=sales.columns, index=range(1, len(sales)+1))
sales1
```

```python
# Using loc to print the index of these columns
sales1.loc[0:7, ["customer_id", "device_id", "item_id"]]
```

```python
# Setting the index of the dataset using the column "locations"
sales.set_index("location", inplace=True)
sales
```

```python
# Display the first five elements in the dataset
sales.head()
```

```python
# Displaying the location of the data set as the index
sales.index
```

```python
# Displaying the list of unique countries with customers
unique_index = sales.index.unique().sort_values()
unique_index
```

```python
# To check the top 10 countries with the most sales
sales.index.value_counts().head(10)
```

`Data visualization of top ten countries with most sales using pandas`

```python
# A barchart of the top ten countries with most sales
sales.index.value_counts().head(10).plot(kind= 'bar',color='black', title='Top ten countries with most sales')
```
