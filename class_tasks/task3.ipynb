**Given the following matrices:**

``` 
Let:    A = 2  -1           B =  1   4  
            0   3               -2   5

            scalars - (c = 2, d = 3)
```

**Question A**

Compute the following:
```
        (cA) = 2 [2     -1]     = [(2x2)  (2x-1)]       = [4    -2]
                 [0      3]       [(2x0)   (3x2)]       = [0     6]

        (dB) = 3 [1      4]     = [(1x3)    (4x3)]       = [3     12]
                 [-2     5]       [(-2x3)   (5x3)]       = [-6    15]

        
        (c + d)A = (2 + 3)A = 2A + 3A  = 2[2     -1]  +  3[2     -1] = [(2x2) (2x-1)] [(3x2)(3x-1)]
                                          [0      3]      [0      3]   [(2x0) (3x2)]  [(3x0)+(3x3)]

        cA + dA =   [4  -2]     +   [6  -3]     =       [10     -5]
                    [0   6]         [0   9]             [0      15]
```

**Question B**

Using the results gotten, verify the scalar Multiplication Properties:

```
Let:    A = 2  -1           B =  1   4  
            0   3               -2   5

            scalars - (c = 2, d = 3)
```

Properties of scalar multiplication


i. Associativity :

```
        (c.d)A = c(dA)
        
    (c.d)A  =   (2x3)[2     -1]     = 6 [2     -1]   =  [(6x2 6x-1)]    =   [12     -6]
                     [0      3]         [0      3]      [(6x0 6x3)]         [0      18]

    c(dA)  =    2 [(3x2)    (3x-1)]   =  2 x [6     -3]     =   [12     -6]
                  [(3x0)     (3x3)]          [0      9]         [0      18]
```


ii.Distributivity over Matrix Addition: 

```
        c(A + B) = cA + cB
        
    c(A + B)  = cA + cB = 2 [2     -1] + 2 [1      4]  = [(2x2 2x-1)]  + [(2x1)    (2x4)] = [6    6]
                            [0      3]     [-2     5]    [(2x0 2x3)]     [(2x-2)   (2x5)]   [-4   16]

    cA + cB =   [4  -2]     +   [2    8]     =     [6      6]
                [0   6]         [-4  10]           [-4    16]
```


iii.Distributivity over Scalar Addition:
```
    (c + d)A = cA  +  dA
    
    (c + d)A = (2 + 3)A =  2A + 3A = 2[2    -1] +  3[2    -1] = [(2x2) (2x-1)] [(3x2)(3x-1)] = [10   -5]
                                      [0     3]     [0     3]   [(2x0) (3x2)]  [(3x0)+(3x3)]   [0    15]

    cA + dA =   [4  -2]     +   [6  -3]     =       [10     -5]
                [0   6]         [0   9]             [0      15]
```

**Question C**

Find the transpose of both matrices:
```

    A^T =    [2  -1 ]       =>       [2      0]
             [0   3 ]                [-1     3]

    B^T =   [1      4]      =>       [1     -2]
```

```
        Let:    A = 2  -1           B =  1   4  
                    0   3               -2   5
```

- Determinant of A:

```
        det(A) = (2x3) - (-1x0) = 6 - 0 = 6
```

- Inverse of A:

```
        A^-1    = 1/6 [3     1]     =   [(3/6)  (1/6)]     =    [1/2    1/6]
                      [0     2]         [(0/6)  (2/6)]          [0      1/3]
```

- Determinant of B:

```
        det(B) = (1x5) - (-2x4) = 5 + 8 = 13
```

- Inverse of B:

```
        B^-1    = 1/13 [5     -4]     =   [(5/13)  (-4/13)]     =   [5/13    -4/13]
                        [2      1]        [(2/13)  (1/13)]          [2/13     1/13]
```
