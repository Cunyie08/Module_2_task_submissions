## **Data Analysis for Electronics Sales** ##



```python
# Importing all required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl as op
 

```

```python
# Loading the dataset
sales = pd.read_excel("electronic_sales.xlsx")
```

```python
# Viewing the dataset
sales 

# Checking the data type for each column
sales.dtypes
```

```python
# Renaming the column names to lower and removing whitespaces
e_sales = sales.rename(columns= lambda x: x.strip().lower().replace(" ", "_"))
e_sales

# Checking the data type for each column
e_sales.dtypes
```

```python
# Creating a revenue column in the dataset
e_sales['revenue'] = e_sales['units'] * e_sales['price']

# Displaying the parameter for the revenue column in thousands
e_sales['revenue(₦)'] = round(e_sales['units'] * e_sales['price'] * 1000,2)
e_sales
```

```python
# Extracting the weekday, month and year from the date column
e_sales['date'] = pd.to_datetime(e_sales['date'])
e_sales['year'] = e_sales['date'].dt.year
e_sales['month'] = e_sales['date'].dt.month_name()
e_sales['week_day'] = e_sales['date'].dt.day_name() 

e_sales.head()


```

## **Working on KPI's** ##

```python
# Total Electronics units sold
total_units_sold = e_sales['units'].sum()
print(f"Total Electronic units sold: {total_units_sold}" )
```

```python
# Total Revenue
total_revenue = e_sales['revenue(₦)'].sum()
print(f"The total revenue generated from electronics is {total_revenue:,}")
```

```python
# Average Revenue per unit 
avrg_revenue_per_unit = round(total_revenue / total_units_sold,2) # using round to approx. up to 2 decimal places
print(f"The Average revenue generated per unit is ₦{avrg_revenue_per_unit:,}")
```

```python
# Number of Products using .nunique() to show the number of unique products
no_of_products = e_sales['products'].nunique()
print(f"Number of products: {no_of_products}")

# List of products sold using .unique() to show the name of the products
name_of_products = e_sales['products'].unique()
print(f"Name of products: {name_of_products}")

```

```python
# Number of Sales Agents using nunique() to show the unique values
no_of_agents = e_sales['sales_agent'].nunique()
print(f"Total number of the Sales Agents: {no_of_agents}")

# Number of Sales Agents using nunique() to show the unique values
name_of_agents = e_sales['sales_agent'].unique()
print(f"Names of the Sales Agents: {name_of_agents}")
```

## **Working on the insights** ##

```python
# Setting the index using the date column
e_sales.set_index('date', inplace=True)
e_sales
```

```python
# Number of unique Year(s) in the data set
e_sales.year.unique()
```

```python
# Display the sales for 2015 only
e_sales_2015 = e_sales[e_sales['year'] == 2015]
e_sales_2015
```

## **Total Revenue by Month** ##

```python
# Using .sort_index() to arrange the output based on the index for 2014
monthly_revenue_2014 = e_sales_2014.groupby('month')['revenue(₦)'].sum().sort_index()
monthly_revenue_2014
```

```python
# Using .sort_index() to arrange the output based on the index for 2015
monthly_revenue_2015 = e_sales_2015.groupby('month')['revenue(₦)'].sum().sort_index()
monthly_revenue_2015
```

```python
# Using .sort_values to arrange the output based on the values for 2014 (from low to high)
monthly_revenue_2014 = e_sales_2014.groupby('month')['revenue(₦)'].sum().sort_values(ascending=True)
print(f"Total monthly revenue in 2014 (ascending order):\n {monthly_revenue_2014}")
```

```python
# Using .sort_values to arrange the output based on the values for 2015 (from low to high)
monthly_revenue_2015 = e_sales_2015.groupby('month')['revenue(₦)'].sum().sort_values(ascending=True)
print(f"Total monthly revenue in 2015 (ascending order):\n {monthly_revenue_2015}")
```

```python
# Using .sort_values to arrange the output based on the values for 2014 & 2015 (from low to high)
total_monthly_revenue = e_sales.groupby('month')['revenue(₦)'].sum().sort_values(ascending=True)
print(f"The overall monthly revenue for 2014 & 2015 (ascending order):\n {total_monthly_revenue}")
```
